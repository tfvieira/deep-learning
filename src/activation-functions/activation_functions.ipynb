```
%tensorflow_version 1.x
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```
# Threshold Activation function
def threshold (x):
    cond = tf.less(x, tf.zeros(tf.shape(x), dtype = x.dtype))
    out = tf.where(cond, tf.zeros(tf.shape(x)), tf.ones(tf.shape(x)))
    return out
```

```
# Plotting Threshold Activation Function
h = np.linspace(-1,1,50)
out = threshold(h)

init = tf.global_variables_initializer()
with tf.Session() as sess:
    sess.run(init)
    y = sess.run(out)
    
plt.xlabel('Activity of Neuron')
plt.ylabel('Output of Neuron')
plt.title('Threshold Activation Function')
plt.plot(h, y)
plt.show()
```

```
# Plotting Sigmoidal Activation function
h = np.linspace(-10,10,50)
out = tf.sigmoid(h)
with tf.Session() as sess:
    sess.run(init)
    y = sess.run(out)

plt.xlabel('Activity of Neuron')
plt.ylabel('Output of Neuron')
plt.title('Sigmoidal Activation Function')
plt.plot(h, y)
plt.grid()
plt.show()
```

```
# Plotting Hyperbolic Tangent Activation function
h = np.linspace(-10,10,50)
out = tf.tanh(h)
with tf.Session() as sess:
    sess.run(init)
    y = sess.run(out)
    
plt.xlabel('Activity of Neuron')
plt.ylabel('Output of Neuron')
plt.title('Hyperbolic Tangent Activation Function')
plt.plot(h, y)
plt.grid()
plt.show()
```

```
# Plotting HardTan Activation function
h = np.linspace(-10,10,50)

def hardtanh(h):
  return tf.maximum(tf.minimum(h,1.),-1.)

out = hardtanh(h)
# out = relu(h)
with tf.Session() as sess:
    sess.run(init)
    y = sess.run(out)
    
plt.xlabel('Activity of Neuron')
plt.ylabel('Output of Neuron')
plt.title('Hard Tanh Activation Function')
plt.plot(h, y)
plt.grid()
plt.show()
```

```
def softsign(x):
  return x/(1 + tf.abs(x))

def softsign_prime(x):
  return 1/((1 + tf.abs(x))**2)

h = np.linspace(-4,4,1000)
out = softsign(h)
out2 = softsign_prime(h)
with tf.Session() as sess:
  sess.run(init)
  y = sess.run(out)
  y2 = sess.run(out2)

plt.plot(h, y)
plt.plot(h, y2)
plt.legend(['SoftSign', 'SoftSign_prime'])
plt.grid()
plt.show()
```

```
# Plotting ReLU Activation function
h = np.linspace(-10,10,50)

def relu(h):
  return tf.maximum(0., h)

# out = tf.nn.relu(h)
out = relu(h)
with tf.Session() as sess:
    sess.run(init)
    y = sess.run(out)
    
plt.xlabel('Activity of Neuron')
plt.ylabel('Output of Neuron')
plt.title('ReLU Activation Function')
plt.plot(h, y)
plt.grid()
plt.show()
```

```
# Threshold softplus function
def softplus (x):
    return tf.log(tf.add(tf.exp(x),1.))

def relu(h):
  return tf.maximum(0., h)


h = np.linspace(-4,4,50)
out1 = softplus(h)
out2 = relu(h)

with tf.Session() as sess:
  sess.run(init)
  y1 = sess.run(out1)
  y2 = sess.run(out2)

plt.plot(h, y1)
plt.plot(h, y2)
plt.legend(['SoftPlus = $\log(1+e^x)$', 'ReLU'])
plt.grid()
plt.show()
```

```
# Plotting Softmax Activation function
h = np.linspace(-5,5,50)
out = tf.nn.softmax(h)
with tf.Session() as sess:
    sess.run(init)
    y = sess.run(out)
    
plt.xlabel('Activity of Neuron')
plt.ylabel('Output of Neuron')
plt.title('Softmax Activation Function')
plt.plot(h, y)
# plt.plot(h, h)
plt.grid()
plt.show()

print(np.sum(y))
```

```
def sigmoid(z):
  return 1. / (1 + np.exp(-z))

def binary_crossentropy(y, y_hat):
  return - (y*np.log(y_hat) + (1 - y) * np.log(1 - y_hat))

def mse(y, y_hat):
  return (y - y_hat)**2

n = 100
z = np.linspace(-3,3,n)
y_pred = sigmoid(z)
plt.figure()#figsize=(16,9))
plt.plot(z,y_pred)#, linewidth=4)
plt.plot(z,binary_crossentropy(1, y_pred))#, linewidth=4)
plt.plot(z,mse(1, y_pred))#, linewidth=4)
plt.legend(['Sigmoid', 'Cross-entropy', 'MSE'])#, prop={'size':28})
plt.grid(True)
plt.show()
```
